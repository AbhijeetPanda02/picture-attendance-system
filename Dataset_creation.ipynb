```python
import cv2
import os

def capture_images(student_id, num_images=200, save_dir="Dataset"):
    # Create the directory path for the student
    student_path = os.path.join(save_dir, student_id)
    os.makedirs(student_path, exist_ok=True)

    # Start the webcam
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open the webcam.")
        return

    print(f"Capturing {num_images} images for {student_id}... Press 'q' to quit early.")
    count = 0

    while count < num_images:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame.")
            break

        # Display the camera feed
        cv2.imshow("Capturing - Press 'q' to quit", frame)

        # Save the captured frame
        img_path = os.path.join(student_path, f"{count + 1}.jpg")
        cv2.imwrite(img_path, frame)
        count += 1
        print(f"Captured image {count}/{num_images}")

        # Wait for a short interval before capturing the next image
        cv2.waitKey(100)

        # Press 'q' to exit early
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Early exit.")
            break

    cap.release()
    cv2.destroyAllWindows()
    print(f"Finished capturing. Images saved in: {student_path}")

# Run the script
if __name__ == "__main__":
    student_id = input("Enter student ID (e.g., 2021CS101): ")
    capture_images(student_id)
```
